```python
# =========================================================
# BRAIN TUMOR CLASSIFICATION USING CNN
# =========================================================

# =========================================================
# 1. IMPORT LIBRARIES
# =========================================================

import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow.keras import layers
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing import image_dataset_from_directory

# =========================================================
# 2. DATASET PATHS
# =========================================================

train_dir = '/content/drive/MyDrive/Training'
test_dir = '/content/drive/MyDrive/Testing'

print("TRAIN FOLDERS:", os.listdir(train_dir))
print("TEST FOLDERS :", os.listdir(test_dir))

# =========================================================
# 3. LOAD DATASET
# =========================================================

train_ds = image_dataset_from_directory(
    train_dir,
    image_size=(300, 300),
    batch_size=32,
    validation_split=0.2,
    subset='training',
    seed=1
)

val_ds = image_dataset_from_directory(
    train_dir,
    image_size=(300, 300),
    batch_size=32,
    validation_split=0.2,
    subset='validation',
    seed=1
)

test_ds = image_dataset_from_directory(
    test_dir,
    image_size=(300, 300),
    batch_size=32,
    shuffle=False
)

# =========================================================
# 4. CLASS NAMES
# =========================================================

class_names = train_ds.class_names

print("\nCLASS NAMES:")
for i, cls in enumerate(class_names):
    print(f"{i} --> {cls}")

# =========================================================
# 5. IMPROVE PERFORMANCE
# =========================================================

AUTOTUNE = tf.data.AUTOTUNE

train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
test_ds = test_ds.cache().prefetch(buffer_size=AUTOTUNE)

# =========================================================
# 6. BUILD CNN MODEL
# =========================================================

model = tf.keras.Sequential([

    layers.Rescaling(1./255, input_shape=(300,300,3)),

    layers.Conv2D(32, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Flatten(),

    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),

    layers.Dense(4, activation='softmax')
])

# =========================================================
# 7. COMPILE MODEL
# =========================================================

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# =========================================================
# 8. MODEL SUMMARY
# =========================================================

model.summary()

# =========================================================
# 9. TRAIN MODEL
# =========================================================

history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=20,
    verbose=1
)

# =========================================================
# 10. EVALUATE MODEL
# =========================================================

test_loss, test_acc = model.evaluate(test_ds)

print("\n==============================")
print(f"TEST ACCURACY : {test_acc:.2%}")
print(f"TEST LOSS     : {test_loss:.4f}")
print("==============================")

# =========================================================
# 11. PLOT ACCURACY
# =========================================================

plt.figure(figsize=(8,5))

plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')

plt.title("Model Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()

plt.show()

# =========================================================
# 12. PLOT LOSS
# =========================================================

plt.figure(figsize=(8,5))

plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')

plt.title("Model Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()

plt.show()

# =========================================================
# 13. SAVE MODEL
# =========================================================

model.save('/content/drive/MyDrive/brain_tumor_cnn.h5')

print("\nModel Saved Successfully!")

# =========================================================
# 14. SINGLE IMAGE PREDICTION
# =========================================================

test_image_path = '/content/drive/MyDrive/Testing/no_tumor/image(100).jpg'

print("\nImage Exists:", os.path.exists(test_image_path))

img = image.load_img(
    test_image_path,
    target_size=(300,300)
)

plt.figure(figsize=(6,6))
plt.imshow(img)
plt.axis('off')
plt.show()

img_array = image.img_to_array(img)
img_array = img_array / 255.0
img_array = np.expand_dims(img_array, axis=0)

prediction = model.predict(img_array)

predicted_class = np.argmax(prediction)

predicted_label = class_names[predicted_class]

confidence = np.max(prediction) * 100

print("\nPrediction Probabilities:")
print(prediction)

print("\nPredicted Tumor Type :", predicted_label)
print("Confidence           :", round(confidence,2), "%")
```
